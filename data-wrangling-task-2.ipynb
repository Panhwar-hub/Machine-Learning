### Interquartile 
In descriptive statistics, the interquartile range is a measure of statistical dispersion, which is the spread of the data. The IQR may also be called the midspread, middle 50%, fourth spread, or Hâ€‘spread. It is defined as the difference between the 75th and 25th percentiles of the data
* I performed numpy subtraction between 75 and 25 percentile of data.

```python
iqr = np.subtract(*np.percentile(data2021, [75, 25]))
```

### Box and Whisker Plot
In a graphic known as a box and whisker, the lower and higher quartiles are represented by the box's left and right sides, respectively. The area within the box corresponds to the interquartile range, which contains fifty percent of the available data. The median is the horizontal line that runs across the middle of the box. On the box plot, the mean may also be represented as a dot or a cross in some instances.
We are able to clearly detect the data location as well as the data spread when using the box and whisker plot, which is one of its advantages.
The skewness and symmetry of the data are both provided by it.
The data outliers are displayed using a box and whisker plot.
In this example we can see how distributed are values of BTC-to-USD each day we can see that the box represents that data is mostly spread between -2000 to 2000 while having some of the data out of range known as outliers.
