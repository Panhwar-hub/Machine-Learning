### Histogram

### Shape of Data 
- Shape of data can be visualized using skewness and Kurtosis values of data 
- I have calculated skewness and Kurtosis for both male and female weights.
I am using scipy module here to find the shape od data, I will be finding skewness of data using `skew()` method from scipy and to find the kurtosis value I will be using `kurtosis()` method of scipy. To find these for only weights of male and female I am only giving one column of data, and to select column I have given `axis=0` which selects column data.

```python
print("Skewness of Female weights",skew(female[:,0], axis=0, bias=True))
print("Kurtosis of Female weights",kurtosis(female[:,0], axis=0, bias=True))

print("Skewness of Male weights",skew(male[:,0], axis=0, bias=True))
print("kurtosis of Male weights",kurtosis(male[:,0], axis=0, bias=True))
```

##### Explanation
    From results we can see that both male and female has positive skewness. From the kurtosis we can see that data is tailed with value of 1.4 for female data and 1.47 for male data.

### BMI Column

### Standardization 
- Standardizing female mesurements data with zscore of each column 

To standardize the data I am using a scipy method that is `zscore` which find z-score of data from the column and then I am dividing each value of column with this z-score, here I am giving `axis=0` to divide with z-score in a row wise manner. To perfrom this action I will read female dataset into a dataframe using `pd.read_csv` which reads a data from csv file and stores in a dataframe.

```python
female_df = pd.read_csv("/kaggle/input/dataset-male-female-body-measurements/nhanes_adult_female_bmx_2020.csv")
```

Data that we read from a csv file does'nt contain BMI column so here I will add BMI valued into a new column of dataframe.

```python
female_df["BMI"] = Col
```

Let's visualize how data looks now in a dataframe, and is the new column added to the dataframe

```python
female_df
```

Here we can see that a new column is added into a dataframe 

##### Standardization
Now to standardize the data I will make a for loop that loops through all the columns present in dataframe and standardize the data, to standardize the data using zscore I am calculating zscore first of each column using `zscore = x-mean(x)/std` with this I get standardized data.

```python
for col in ["BMXWT","BMXHT","BMXARML","BMXLEG","BMXARMC","BMXHIP","BMXWAIST","BMI"]:
    female_df[col] = (female_df[col]-female_df[col].mean())/female_df[col].std()
```

```python
female_df
```

After implementing above loop we can see that all the values in data are changed now to same scaled standard deviation, hence we can say that data is now standardized
