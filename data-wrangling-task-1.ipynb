* Defining three lists to store names, heights, and weights of friends.

```python
names = ["James","Charlie","bonnie","riche","dormet","nancy"]
heights = [1.6,1.77,1.59,1.65,1.72,1.8]
weights = [62,73,59,90,68,69]
```

* Making a new list which will have BMI for each person from names list above. BMI will be calculated using heights and weights provided in above lists.

```python
BMI= []
for v in range(0,len(names)): 
    BMI =BMI+ [weights[v]/(heights[v]*heights[v])]
```

* Importing Matplotlib module to plot BMI for each person.
* Plotted a bar plot having names of person on horizontal scale, and corresponding BMI in vertical column.

```python
import matplotlib.pyplot as plt
plt.bar(names, BMI)

plt.title("My friend's BMIs")
plt.show()
```

* A category is defined based on Calculated values of BMI.
* Below is a function defined to provide a category according to BMI of a person.
* initially empty string named cat and storing category in it.

```python
cat = ""
def BMI_CAT(bmi):
    if (bmi<18.5):
        cat="Lower weight"
    elif(bmi > 18.5 and bmi<22.9):
        cat = "Normal"
    elif (bmi > 23 and bmi <24.9):
        cat = "Overweight"
    elif (bmi > 25 and bmi < 29.9):
        cat = "Obese"
    else:
        cat = "Severly overweight"
    return cat    
```

* Now using formatted print feature of python to print names of people with their corresponding BMI and it's category.

```python
for i in range(0,len(names)):
    print(f"{names[i]} has bmi of {BMI[i]} which is {BMI_CAT(BMI[i])} .")
```

### Benifits and Limitation of BMI According Medical
BMI isÂ  A quick, low-cost, and non-invasive substitute for measuring body fat is BMI. Unlike other approaches, BMI just considers a person's height and weight, and with the right tools, any individual can have their BMI regularly assessed and computed with fair precision.
At the population level, BMI serves as a helpful health indicator. When determining your risk of developing a disease, the allocation of fat on your body matters more than the total amount. The waist circumference is therefore considered to be a more accurate indicator of health risk than the BMI.
The link between BMI and body fat can be influenced by elements including age, sex, ethnicity, and muscle mass. Additionally, the BMI does not discriminate between extra fat, muscle, or bone mass, nor does it show how fat is distributed among people.

### Possible missuses of BMI
Surprisingly, a Belgian founded the BMI. Adolphe Quetelet collected data on men's heights and weights at different ages in the 1830s. From this study, which he believed would help him estimate the 'average' man, he developed the Quetelet formula, now known as the BMI, which is derived by dividing a person's weight in kilogrammes by their height in metres squared. It may be surprising to learn that its optimal usage in modern clinical practise requires additional considerations to be considered. These considerations are overlooked, though. It was created for Belgians, not Bangladeshis, Botswanans, or boxers. If its function is to identify if someone is fat and needs action to restore health or minimise risk of illness, then people who use it should understand it better than most do.
