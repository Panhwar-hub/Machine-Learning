Predicting number of orders from pakisatan's largest E-commerce dataset. For prediction there are many of classification algorithms that can be used. Here  I am using linear classification model known as regression model.

```python
#importing libraries
import pandas as pb 
import matplotlib.pyplot as plt
import numpy as np 
# load dataset.
data= pb.read_csv('../input/pakistans-largest-ecommerce-dataset/Pakistan Largest Ecommerce Dataset.csv')
data.info()
```

```python
#creating a sub dataset with needed columns and negleting others.
data_sub = data[['Month', 'item_id','Year','qty_ordered', 'grand_total','price','discount_amount']]
#data_sub
```

```python
# removing NaN and null values from dataset
data_sub.dropna(how='all', inplace=True)
data_sub.sort_index(ascending=False)
```

```python
#removing NUll from remaning columns.
data_sub.dropna(inplace=True)
```

```python
# checking no null value remains.
print(data_sub.isnull().sum())
```

```python
# for model prediction creating labels and features.
y=data_sub['qty_ordered']
x=data_sub.drop('qty_ordered', axis=1)
```

```python
#train test and split the data
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest=train_test_split(x,y, test_size=0.2)
```

```python
# check the dimentions of the splited data.
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

Predicting model using logistic regression 

```python
# fit model using logistic regresion.
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(xtrain, ytrain)
```

```python
# model prediction for xtest data.
prediction=model.predict(xtest)
prediction
print('Accuracy of model is:',model.score(xtest, ytest))
```

```python
# model accuracy and classification.
from sklearn.metrics import classification_report
print(classification_report(ytest,prediction))
```
